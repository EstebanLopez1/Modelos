```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
import numpy as np
import seaborn as sns
plt.style.use("classic")

df = pd.read_csv (r'C:\Users\esteb\OneDrive\Escritorio\Maestr√≠a Analytics\Vacacional\Oil.csv', skiprows=2)
columnas=["Mes", "Oil"]
df.columns=columnas
df["Mes"] = pd.to_datetime(df['Mes'])
df.set_index('Mes', inplace=True)
df.head()
```

```python
df.plot(figsize=(8, 5), grid=True, linewidth=3, fontsize=15, title="Serie de Timepo Petr√≥leo")
plt.ylabel("Inter√©s de B√∫squeda")
plt.axhline(df["Oil"].mean(), color ="red", linestyle="dashed", label="Promedio")
plt.legend()
plt.show()
```

```python
result = adfuller(df["Oil"])
result[1]
```

```python
media_movil=df["Oil"].rolling(12).mean()
media_movil.plot(figsize=(8,5), linewidth=3, fontsize=15)
plt.ylabel("Media Movil Inter√©s de B√∫squeda")
plt.show()
```

```python
df_dif=df["Oil"].diff()
df_dif.plot(figsize=(8,5), linewidth=2, fontsize=15)
plt.xlabel('Year', fontsize=20);
```

```python
result = adfuller(df_dif.dropna())
result[1]
```

```python
plt.figure(figsize=(8,5))
pd.plotting.autocorrelation_plot(df,linewidth=3, color='brown')
plt.title('Autocorrelacion - Oil')
plt.xlabel('Rezago (lag) en meses')
plt.xticks(np.arange(1,12), rotation=90, fontsize=6);
```
